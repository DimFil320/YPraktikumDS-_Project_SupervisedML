```python
# –∑–∞–≥—Ä—É–∂–∞–µ–º –¥–∞—Ç–∞—Å–µ—Ç—ã
# –≤ –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö —Ñ–∞–π–ª–∞—Ö —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å –∑–Ω–∞—á–µ–Ω–∏–π ‚Äî —Ç–æ—á–∫–∞ —Å –∑–∞–ø—è—Ç–æ–π, –∞ —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å –¥–µ—Å—è—Ç–∏—á–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π ‚Äî –∑–∞–ø—è—Ç–∞—è
market = pd.read_csv('/datasets/market_file.csv')
m_money = pd.read_csv('/datasets/market_money.csv')
m_time = pd.read_csv('/datasets/market_time.csv')
money = pd.read_csv('/datasets/money.csv', sep=';', decimal=",")
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏–∏ –∏ –æ—á–∏—Å—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö –≤ DataFrame
def clean_dataframe(df):
    # –ø—Ä–∏–≤–µ–¥–µ–º –∏–º–µ–Ω–∞ –∫–æ–ª–æ–Ω–æ–∫ –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É, –∑–∞–º–µ–Ω–∏–º –ø—Ä–æ–±–µ–ª—ã –Ω–∞ –ø–æ–¥—á–µ—Ä–∫–∏–≤–∞–Ω–∏—è
    df.columns = df.columns.str.lower().str.replace(' ', '_')
    
    # –ø—Ä–∏–≤–µ–¥–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –≤ –∫–æ–ª–æ–Ω–∫–∞—Ö –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É
    df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)
    
    return df

# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã–≤–æ–¥–∞ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ DataFrame —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ info() –∏ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è —Ç–∞–±–ª–∏—á–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Å –ø–æ–º–æ—â—å—é display()
def display_info(df_list):
    for df in df_list:
        print("Info –æ DataFrame:")
        display(df.info())
        print("–¢–∞–±–ª–∏—á–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è DataFrame:")
        display(df)

market = clean_dataframe(market)
m_money = clean_dataframe(m_money)
m_time = clean_dataframe(m_time)
money = clean_dataframe(money)

display_info([market, m_money, m_time, money])
```

```python
def check_duplicates(df_list):
    duplicates_info = {}
    for idx, df in enumerate(df_list):
        duplicates = df.duplicated().any()
        if duplicates:
            duplicates_info[f'DF_{idx+1}'] = "–î—É–±–ª–∏–∫–∞—Ç—ã –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É—é—Ç"
        else:
            duplicates_info[f'DF_{idx+1}'] = "–î—É–±–ª–∏–∫–∞—Ç—ã –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç"
    return duplicates_info

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –≤ –∫–∞–∂–¥–æ–º –∏–∑ –∑–∞–≥—Ä—É–∂–µ–Ω–Ω—ã—Ö DataFrame
duplicates_info = check_duplicates([market, m_money, m_time, money])

# –í—ã–≤–æ–¥ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ –Ω–∞–ª–∏—á–∏–∏ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤
for df_name, info in duplicates_info.items():
    print(f"{df_name}: {info}")
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∑–Ω–∞—á–µ–Ω–∏–π
def plot_distribution(data, column, bins=None):
    plt.figure(figsize=(8, 6))
    plt.hist(data[column], bins=bins, edgecolor='k', alpha=0.7)
    plt.title(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π '{column}'", fontsize=15)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞', fontsize=12)
    plt.grid(True)
    plt.show()

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –¥–ª—è –∫–∞–∂–¥–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π, –∫—Ä–æ–º–µ 'id'
for column in market.columns:
    if column != 'id' and market[column].dtype in ['int64', 'float64']:
        plot_distribution(market, column)
    elif column != 'id':
        print(f"–ü–µ—Ä–µ–º–µ–Ω–Ω–∞—è '{column}' –Ω–µ —è–≤–ª—è–µ—Ç—Å—è —á–∏—Å–ª–æ–≤–æ–π –∏ –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–∞.")
```

## –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ —Ç–∞–±–ª–∏—Ü

```python
m_money_pivot = (m_money
                      .pivot_table(index='id', 
                                   columns='–ø–µ—Ä–∏–æ–¥', 
                                   values='–≤—ã—Ä—É—á–∫–∞', 
                                   aggfunc='sum')
                      .rename(columns={'–ø—Ä–µ–¥—ã–¥—É—â–∏–π_–º–µ—Å—è—Ü': '–≤—ã—Ä—É—á–∫–∞_–ø—Ä–µ–¥',
                                       '–ø—Ä–µ–ø—Ä–µ–¥—ã–¥—É—â–∏–π_–º–µ—Å—è—Ü': '–≤—ã—Ä—É—á–∫–∞_–ø—Ä–µ–ø—Ä–µ–¥',
                                       '—Ç–µ–∫—É—â–∏–π_–º–µ—Å—è—Ü': '–≤—ã—Ä—É—á–∫–∞_—Ç–µ–∫'})
                      .dropna()
                      .reset_index())
m_money_pivot.head()
```

```python
m_time_pivot = (m_time
                     .pivot_table(index='id', 
                                  columns='–ø–µ—Ä–∏–æ–¥', 
                                  values='–º–∏–Ω—É—Ç', 
                                  aggfunc='sum')
                     .fillna(0)
                     .rename(columns={'–ø—Ä–µ–¥—ã–¥—É—â–∏–π_–º–µ—Å—è—Ü': '–º–∏–Ω—É—Ç_–ø—Ä–µ–¥',
                                      '—Ç–µ–∫—É—â–∏–π_–º–µ—Å—è—Ü': '–º–∏–Ω—É—Ç_—Ç–µ–∫'})
                     .reset_index())
```

–ö–æ–≥–¥–∞ —É –Ω–∞—Å –±–∏–Ω–∞—Ä–Ω–∞—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è, —Ç–æ –º–µ—Ç—Ä–∏–∫–∏ presicion, recall, f1 —Å—á–∏—Ç–∞—é—Ç—Å—è (–∏ –ø–æ–¥–±–∏—Ä–∞–µ—Ç—Å—è –≤ GS) —Ç–æ–ª–∫–æ –ø–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–º—É –∫–ª–∞—Å—Å—É.  –ü–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Å - —ç—Ç–æ —Ç–æ—Ç –∫–ª–∞—Å—Å –∫–æ—Ç–æ—Ä—ã–π –Ω–∞—Å –±–æ–ª—å—à–µ –∏–Ω—Ç–µ—Ä–µ—Å—É–µ—Ç, –ø–æ –ø–æ—Å—Ç–∞–Ω–æ–≤–∫–µ –∑–∞–¥–∞—á–∏ –≤–∏–¥–Ω–æ —á—Ç–æ —ç—Ç–æ "—Å–Ω–∏–∑–∏–ª–∏ –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å". –ê —Ç–∞–∫ –Ω–µ–ø–æ–Ω—è—Ç–Ω–æ –∫–∞–∫ —É —Ç–µ–±—è –∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞–ª–æ—Å—å
    

```python
# —Å–æ–∑–¥–∞–¥–∏–º –º–∞—Ç—Ä–∏—Ü—ã –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π '–ø–æ–∫—É–ø–∞—Ç–µ–ª—å—Å–∫–∞—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å'
X = total_market.drop(['id', '–ø–æ–∫—É–ø–∞—Ç–µ–ª—å—Å–∫–∞—è_–∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å'], axis=1)
y = total_market['–ø–æ–∫—É–ø–∞—Ç–µ–ª—å—Å–∫–∞—è_–∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å']

# —Ä–∞–∑–¥–µ–ª–∏–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ–±—É—á–∞—é—â–∏–π –∏ —Ç–µ—Å—Ç–æ–≤—ã–π –Ω–∞–±–æ—Ä—ã
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    stratify=y,
                                                    test_size=TEST_SIZE,
                                                    random_state=RANDOM_STATE) # –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞–¥–∞–≤–∞–ª–∏ –≤ —Å–∞–º–æ–º –Ω–∞—á–∞–ª–µ
# –æ–ø—Ä–µ–¥–µ–ª–∏–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –∏ —á–∏—Å–ª–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
cat_columns = X_train.select_dtypes('object').columns
num_columns = X_train.select_dtypes('float').columns

# —Å–æ–∑–¥–∞–¥–∏–º –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
cat_pipe = Pipeline(
    [
        (
            'simpleImputer_before_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'encoder', 
            OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=np.nan)),
        (
            'simpleImputer_after_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)

# —Å–æ–∑–¥–∞–¥–∏–º –æ–±—â–∏–π –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø—Ä–µ–æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö
data_pipe = ColumnTransformer(
    [
        (
            'cat', 
            cat_pipe,
            cat_columns
        ),
        (
            'num', 
            StandardScaler(), 
            num_columns
        ),
    ]
)

#  —Å–æ–∑–¥–∞–¥–∏–º –ø–∞–π–ø–ª–∞–π–Ω —Å –º–æ–¥–µ–ª—å—é —Ä–µ—à–∞—é—â–µ–≥–æ –¥–µ—Ä–µ–≤–∞
final_pipe = Pipeline(
    [
        ('preprocessor', data_pipe),
        ('classifier', DecisionTreeClassifier(random_state=RANDOM_STATE))
    ]
)
# –æ–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å
final_pipe.fit(X_train, y_train)

# –ø–æ–ª—É—á–∞–µ–º –ø—Ä–æ–≥–Ω–æ–∑—ã –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–º –Ω–∞–±–æ—Ä–µ –∏ –æ—Ü–µ–Ω–∫–∞ –º–µ—Ç—Ä–∏–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏
y_test_pred = final_pipe.predict_proba(X_test)
print(f'–ú–µ—Ç—Ä–∏–∫–∞ ROC-AUC –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ: {roc_auc_score(y_test, y_test_pred[:, 1])}')
```

ROC-AUC –≤ –∑–Ω–∞—á–µ–Ω–∏–∏ 0.71 –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –º–æ–¥–µ–ª—å –¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É–µ—Ç –Ω–µ–ø–ª–æ—Ö—É—é —Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç—å —Ä–∞–∑–ª–∏—á–∞—Ç—å –º–µ–∂–¥—É –∫–ª–∞—Å—Å–∞–º–∏ "–ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω—ã–µ –ø–æ–∫—É–ø–∞—Ç–µ–ª–∏" –∏ "–Ω–µ–ø–æ–∫—É–ø–∞—Ç–µ–ª–∏", –æ–¥–Ω–∞–∫–æ, –∑–Ω–∞—á–µ–Ω–∏–µ 0.70 –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –µ—Å—Ç—å –Ω–µ–∫–æ—Ç–æ—Ä–æ–µ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ –¥–ª—è —É–ª—É—á—à–µ–Ω–∏—è. –ú–æ–¥–µ–ª—å –º–æ–∂–µ—Ç –±—ã—Ç—å –¥–æ—Ä–∞–±–æ—Ç–∞–Ω–∞ –¥–ª—è —É–ª—É—á—à–µ–Ω–∏—è –µ–µ —Ç–æ—á–Ω–æ—Å—Ç–∏ –∏ —Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç–∏ —Ä–∞–∑–ª–∏—á–µ–Ω–∏—è –∫–ª–∞—Å—Å–æ–≤.

–¢–æ–≥–¥–∞   –Ω–∞—Ä—É—à–µ–Ω–∏—è –ª–æ–≥–∏–∫–∏ –º–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–∏—è –Ω–µ—Ç, –æ–∫–µ. –•–æ—Ç—è —è –≤—Å—ë —Ä–∞–≤–Ω–æ —É–±—Ä–∞–ª –±—ã 
    
    
    
    # –æ–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å
    final_pipe.fit(X_train, y_train)

    # –ø–æ–ª—É—á–∞–µ–º –ø—Ä–æ–≥–Ω–æ–∑—ã –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–º –Ω–∞–±–æ—Ä–µ –∏ –æ—Ü–µ–Ω–∫–∞ –º–µ—Ç—Ä–∏–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏
    y_test_pred = final_pipe.predict_proba(X_test)
    print(f'–ú–µ—Ç—Ä–∏–∫–∞ ROC-AUC –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ: {roc_auc_score(y_test, y_test_pred[:, 1])}')
    
    
    
–í—Å—ë-—Ç–∞–∫–∏ —ç—Ç–æ –ø—Ä–æ–µ–∫—Ç, –∞ –Ω–µ —á–µ—Ä–Ω–æ–≤–∏–∫    
    
    
    
    
–ê –≤–æ-–≤—Ç–æ—Ä—ã—Ö —Ç—ã –Ω–∏–∫–∞–∫ –Ω–µ –æ–±–æ—Å–Ω–æ–≤–∞–ª –º–µ—Ç—Ä–∏–∫—É roc-auc ( –¢—É—Ç —Å—Ç–æ–∏–ª–æ –Ω–∞–ø–∏—Å–∞—Ç—å —á–µ–º –û–Ω–∞ —Ö–æ—Ä–æ—à–∞ –≤ –æ—Ç–ª–∏—á–∏–µ –æ—Ç –¥—Ä—É–≥–∏—Ö –º–µ—Ç—Ä–∏–∫, –∏–ª–∏ —É–≤—è–∑–∞—Ç—å –µ—ë —Å –±–∏–∑–Ω–µ—Å-–∑–∞–¥–∞—á–µ–π)     

```python
# –æ–ø—Ä–µ–¥–µ–ª–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è –Ω–∞—Å—Ç—Ä–æ–π–∫–∏ –º–æ–¥–µ–ª–µ–π
# –¥–ª—è LogisticRegression
param_grid = {
    'preprocessor__cat__encoder': [OneHotEncoder(drop='first')],
    'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler()],
    'classifier__C': range(1, 5),
    'classifier__penalty': ['l2', 'l1'],
    'classifier': [LogisticRegression(random_state=RANDOM_STATE, solver='liblinear', max_iter=1000)]
}


# –¥–ª—è KNeighborsClassifier
param_knn = {
    'preprocessor__cat': [OneHotEncoder(drop='first')],
    'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler()],
    'classifier__n_neighbors': [2, 5, 10, 25, 50],
    'classifier': [KNeighborsClassifier()]
}

# –¥–ª—è DecisionTreeClassifier
param_dtc = {
    'preprocessor__cat__encoder': [OrdinalEncoder(), OneHotEncoder(drop='first')],
    'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler()],
    'classifier__max_depth': range(2, 11),
    'classifier__min_samples_leaf': range(2, 10),
    'classifier__min_samples_split': range(2, 10),
    'classifier': [DecisionTreeClassifier(random_state=RANDOM_STATE)]
}

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è SVC
param_svc = {
    'preprocessor__cat__encoder': [OrdinalEncoder(), OneHotEncoder(drop='first')],
    'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler()],
    'classifier__C': [10**-2, 10**-1, 10**0, 10**1, 10**2],
    'classifier': [SVC(probability=True, random_state=RANDOM_STATE)]
}
```

```python
# –ø–æ–¥–±–∏—Ä–∞–µ–º –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –º–æ–¥–µ–ª–µ–π —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º GridSearchCV
params = [param_grid, param_knn, param_dtc, param_svc]
grid = GridSearchCV(final_pipe, 
                    params, 
                    cv=5, 
                    n_jobs=-1,
                    scoring='roc_auc')
grid.fit(X_train, y_train)
```

```python
# –≤—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
print('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –∏ –µ—ë –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:\n\n', grid.best_estimator_)
print('–ú–µ—Ç—Ä–∏–∫–∞ ROC-AUC –¥–ª—è –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏:\n', grid.best_score_)
```

```python
# —Å–æ—Ö—Ä–∞–Ω—è–µ–º –Ω–∞—à –ª—É—á—à–∏–π –ø–∞–π–ø–ª–∞–π–Ω
best_pipe = grid.best_estimator_
# –¥–æ—Å—Ç—É–ø –∫ –ª–∏–Ω–µ–π–Ω–æ–π –º–æ–¥–µ–ª–∏
linear_model = best_pipe.named_steps['classifier']

# –ø–æ–¥–≥–æ—Ç–æ–≤–ª–∏–≤–∞–µ–º –¥–∞–Ω–Ω—ã–µ
X_train_prep = best_pipe.named_steps['preprocessor'].fit_transform(X_train)
X_test_prep = best_pipe.named_steps['preprocessor'].transform(X_test)

explainer = shap.Explainer(linear_model.predict_proba, X_train_prep)
# –≤—ã—á–∏—Å–ª—è–µ–º SHAP –∑–Ω–∞—á–µ–Ω–∏—è
shap_values = explainer(X_test_prep)
```
